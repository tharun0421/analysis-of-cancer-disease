# Machine Learning
## RandomForest Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import log_loss, f1_score
from sklearn.model_selection import cross_val_score
import numpy as np
acc_dict = {}
# create the data
X = df.drop('Level',axis = 1)
y = df['Level']
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
from sklearn.ensemble import  RandomForestClassifier
from sklearn.metrics import accuracy_score, log_loss, f1_score, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt

# Get all unique classes from the original 'Level' column
all_classes = np.unique(df['Level'])

# create model
model = RandomForestClassifier()

# fit the data in the model
model.fit(X_train, y_train)
y_pred_randomF = model.predict(X_test)
y_prob_randomF = model.predict_proba(X_test)

# calculate accuracy score
accuracy = accuracy_score(y_test, y_pred_randomF)
print('Accuracy score : ', accuracy*100)

# calculate log loss
log_loss_val = log_loss(y_test, y_prob_randomF, labels=all_classes)
acc_dict['RFC_log_loss'] = log_loss_val

# calculate f1 score
f1_score_val = f1_score(y_test, y_pred_randomF, average='weighted')
acc_dict['RFC_F1_Score'] = f1_score_val

# visualize confusion matrix
plt.imshow(np.log(confusion_matrix(y_test,y_pred_randomF)),cmap = 'Reds',interpolation = 'nearest')
plt.ylabel('True')
plt.xlabel('Predicted')
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score

precision = precision_score(y_test, y_pred_randomF, average='weighted')
recall = recall_score(y_test, y_pred_randomF, average='weighted')
accuracy = accuracy_score(y_test, y_pred_randomF)

print('Precision: ', precision)
print('Recall: ', recall)
print('Accuracy: ', accuracy)

```

## KNeighbourClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
# to find the best k
score = 0
scoresi, highscore, bestk = 0, 0, 0

for k in range(3,12):
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train)
    score = scores.mean()
    if score>highscore:
        highscore = score
        bestk = k
print('Best k is {} with score {}'.format(bestk, highscore))

knn = KNeighborsClassifier(n_neighbors=bestk)
knn.fit(X_train,y_train)
# prediction
y_predict = knn.predict(X_test)
print('Accuracy score : ',accuracy_score(y_test,y_predict)*100)

# Replace 'high' with 1 before calculating log loss and f1 score
y_test_numeric = y_test.replace('high', '1').astype(int)
y_predict_numeric = pd.Series(y_predict).replace('high', '1').astype(int)

acc_dict['KNN_log_loss'] = log_loss(y_test_numeric, y_predict_numeric)
acc_dict['KNN_F!1_Score'] = f1_score(y_test_numeric, y_predict_numeric,average='weighted')

# prediction visualization
plt.imshow(np.log(confusion_matrix(y_test_numeric,y_predict_numeric)),cmap = 'viridis',interpolation = 'nearest')
plt.ylabel('True')
plt.xlabel('Predicted')
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score

# Calculate precision
precision = precision_score(y_test, y_predict, average='weighted')

# Calculate recall
recall = recall_score(y_test, y_predict, average='weighted')

# Calculate accuracy
accuracy = accuracy_score(y_test, y_predict)

print('Precision: {:.2f}'.format(precision))
print('Recall: {:.2f}'.format(recall))
print('Accuracy: {:.2f}'.format(accuracy))
```
